```r
library('tidyverse')
wind_turbine <- readr::read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv')
```

Each row in the table represents a single turbine.

Make a bar chart 沒 **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
glimpse(wind_turbine)
```

```r
var = wind_turbine$province_territory |>
    fct_infreq()

p = ggplot(wind_turbine, aes(y = var)) + geom_bar()
p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
# wind_turbine$manufacturer

var2 = wind_turbine$manufacturer |>
    fct_lump_n(n=5) |>
    fct_infreq() 
    # fct_relevel('Other', after = Inf)
    
p = ggplot(wind_turbine, aes(x = var2)) + geom_bar(fill="coral")
p
```
